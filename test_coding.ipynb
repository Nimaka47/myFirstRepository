```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 10)
y = 3 * np.random.rand(np.size(x)) + 2 * x ** 2 + 10

order = 2
k = np.polyfit(x, y, order)
Y = np.zeros(np.size(y))
for i in range(order + 1):
        Y += k[i] * x ** (order - i)
```

```python
print(*range(order+1))
print(k)
```

```python
plt.plot(x, Y)
plt.scatter(x, y)

plt.ylabel('some numbers')
plt.xlabel('some numbers')
plt.show()
```
